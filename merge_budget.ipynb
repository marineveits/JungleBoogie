```python
from __future__ import print_function

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image

%matplotlib inline
```

```python
# Open budgets data
budget = pd.read_csv('budget.csv')
budget.head()
```

```python
budget=budget.drop_duplicates(subset='Movie')
budget.info()
```

```python
# Open Boxoffice movie data
box=pd.read_csv('movies_data_to_merge.csv')
box.head()
```

```python
# Merge dataframes
movies_budg = pd.merge(box,
                 budget,
                 on='Movie', how='left')
movies_budg.head()
```

```python
movies_budg.to_csv('moviebudget.csv',  sep=',', encoding='utf-8')
```

```python
df = pd.read_csv('moviebudget_merged.csv')
df.head()
```

```python
#Create Dummy Variable for Top 5 Composers
top_15 = []
for person in df['composer']:
    if (person == 'John Williams' or person == 'Hans Zimmer' or person == 'James Newton Howard' or person == 'Danny Elfman' or person == 'Alan Silvestri'
       or person == 'Michael Giacchino' or person == 'James Horner' or person == 'John Powel' or person == 'John Debney' or person == 'Christophe Beck'
       or person == 'Thomas Newman' or person == 'Howard Shore' or person == 'Harry Gregson' or person == 'Jerry Goldsmith' or person == 'Brian Tyler'):
        top_15.append(int(1))
    else:
        top_15.append(int(0))
```

```python
#Adding Dummy Variable to DF
df['top_15'] = top_15

#Changing Column Names
df.columns = ['adj_rank', 'budget', 'composer', 'domestic_rank', 'dom_total_gross', 'movie_title', 'runtime', 'year', 'yearly_rank', 'top_5']
```
