```python
from __future__ import print_function

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image

%matplotlib inline
```

```python
# Open IMDB data
imdb = pd.read_csv('merged_imdb.csv')
imdb.head()
```

```python
# Open merged budget boxoffice data
boxof = pd.read_csv('moviebudget_merged.csv')
boxof.head()
```

```python
# Merge the dataframes
movies_all = pd.merge(boxof,
                 imdb,
                 on='movie', how='left')
movies_all.head()
```

```python
movies_all.describe()
```

```python
movies_all.to_csv('final_movies.csv',  sep=',', encoding='utf-8')
```
